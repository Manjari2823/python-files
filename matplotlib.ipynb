```python
import matplotlib.pyplot as plt

print("Matplotlib Imported Successfully")
```

```python
x = [1,2,3,4,5]
y = [10,20,15,30,25]

plt.plot(x,y)

plt.title("Line Plot")

plt.xlabel("X Axis")

plt.ylabel("Y Axis")

plt.show()
```

```python
x=[1,2,3,4,5]
y=[5,10,15,20,25]

plt.plot(x,y,marker="o")

plt.title("Line Plot with Marker")

plt.show()
```

```python
plt.plot(x,y,color="red",linewidth=3)

plt.title("Colored Line")

plt.show()
```

```python
x=[1,2,3,4]

sales1=[20,25,30,35]

sales2=[15,18,22,30]

plt.plot(x,sales1,label="Product A")

plt.plot(x,sales2,label="Product B")

plt.legend()

plt.show()
```

```python
subjects=["Python","Java","C++","SQL"]

students=[50,40,30,60]

plt.bar(subjects,students)

plt.title("Bar Chart")

plt.show()
```

```python
plt.barh(subjects,students)

plt.show()
```

```python
x=[5,10,15,20,25]

y=[2,4,6,8,10]

plt.scatter(x,y)

plt.title("Scatter Plot")

plt.show()
```

```python
x=[5,10,15,20,25]

y=[2,4,6,8,10]

plt.scatter(x,y)

plt.title("Scatter Plot")

plt.show()
```

```python
marks=[80,82,85,90,91,92,95,98,75,70,68]

plt.hist(marks,bins=5)

plt.title("Histogram")

plt.show()
```

```python
languages=["Python","Java","C++","SQL"]

students=[40,30,20,10]

plt.pie(students,labels=languages,autopct="%1.1f%%")

plt.title("Pie Chart")

plt.show()
```

```python
months=["Jan","Feb","Mar","Apr","May"]

sales=[100,120,150,130,170]

plt.plot(months,sales,marker="o")

plt.title("Monthly Sales")

plt.xlabel("Months")

plt.ylabel("Sales")

plt.grid()

plt.show()
```
