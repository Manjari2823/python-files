```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

print("Seaborn Version:", sns.__version__)
```

```python
tips = sns.load_dataset("tips")

tips.head()
```

```python
tips.info()
```

```python
tips.describe()
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip")

plt.title("Scatter Plot")

plt.show()
```

```python
sns.lineplot(data=tips,x="size",y="total_bill")

plt.show()
```

```python
sns.barplot(data=tips,x="day",y="total_bill")

plt.show()
```

```python
sns.countplot(data=tips,x="day")

plt.show()
```

```python
sns.histplot(data=tips,x="total_bill",bins=10)

plt.show()
```

```python
sns.kdeplot(data=tips,x="total_bill",fill=True)

plt.show()
```

```python
sns.boxplot(data=tips,x="day",y="total_bill")

plt.show()
```

```python
sns.violinplot(data=tips,x="day",y="tip")

plt.show()
```

```python
sns.stripplot(data=tips,x="day",y="tip")

plt.show()
```

```python
sns.swarmplot(data=tips,x="day",y="tip")

plt.show()
```

```python
sns.pairplot(tips)

plt.show()
```

```python
corr=tips.select_dtypes(include="number").corr()

corr
```

```python
sns.heatmap(corr,annot=True,cmap="Blues")

plt.show()
```

```python
sns.scatterplot(data=tips,
                x="total_bill",
                y="tip",
                hue="sex")

plt.show()
```

```python
sns.set_style("darkgrid")

sns.lineplot(data=tips,x="size",y="tip")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(data=tips,x="day",y="tip")

plt.show()
```

```python
sns.histplot(data=tips,x="tip")

plt.savefig("seaborn_plot.png")

plt.show()
```

```python
sns.boxplot(data=tips,
            x="sex",
            y="tip")

plt.title("Tips by Gender")

plt.show()
```
